```python
import cv2

# Ch·∫°y l·ªánh predict.py b·∫±ng magic command ! (ch·ªâ √°p d·ª•ng trong Colab)
!python tracking_vehicle_counting.py model=yolov8l.pt source="test.mp4"
# yolov8l.pt
# ƒê∆∞·ªùng d·∫´n ƒë·∫øn video ƒë·∫ßu ra
output_video_path = "output/output.mp4"

# T·∫°o video writer
cap = cv2.VideoCapture(output_video_path)
frame_wid = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_hyt = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
writer = cv2.VideoWriter("output.mp4", fourcc, fps, (frame_wid, frame_hyt))

# Ghi khung h√¨nh t·ª´ video ƒë·∫ßu ra v√†o video ƒë·∫ßu ra cu·ªëi c√πng
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    writer.write(frame)

# Gi·∫£i ph√≥ng t√†i nguy√™n
cap.release()
writer.release()
cv2.destroyAllWindows()
```
